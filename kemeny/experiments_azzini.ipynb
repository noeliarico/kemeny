```python
def measure(n, seed, rep = 3, om = None):
    
    # n
    # seed
    # incertidumbre1
    # incertidumbre2
    # incertidumbre3
    # incertidumbre4
    # incertidumbre5
    # hay que mirar tb la media 
    # RESULTS full
    # RESULTS borda
    # Kemeny distance
    # Borda distance
    # winners
    
    if om is None:
        np.random.seed(seed)
        om = init.create_random_om(n, 10) 
    else:
        seed = None
    
    results = np.array([n, seed])
    results = np.append(results, [
               incertidumbre1(om),
               incertidumbre2(om),
               incertidumbre3(om),
               incertidumbre4(om),
               incertidumbre5(om)])
    
    # The first time is slower because it need to be compile so do it random
    np.random.seed(123)
    compileom = init.create_random_om(4, 10) 
    dfs1(compileom, float('inf'))
    ##########################################################################
      
    print("Execution without init...")    
    # Execution time without init distance ###################################
    extime = 0
    for i in range(rep):
        start = time.time()
        sol1 = dfs1(om, float('inf'))
        end = time.time()
        extime += (end-start)
    print(extime/rep)
    results = np.append(results, extime/rep) 
    print(sol1)
    
    a = np.arange(om.shape[0]).tolist() # the ranking
    b = sol1[0].tolist() # the indexes
    sol = np.array([b.index(x) if x in b else None for x in a])
    dk = dRP(sol, om)
    print("Obtained distance is {}".format(dk))
    print(sol)
    
    # Calculate borda count and the distance to the profile
    r = ranking_to_lineal(borda(om))
    db = dRP(r, om)
    print("Execution with borda... (distance = {})".format(db))
    print(r)
    #print(b)
    # Execution time using the distance given by Borda Count
    extime = 0
    for i in range(rep):
        start = time.time()
        sol2 = dfs1(om, db)
        end = time.time()
        extime += (end-start)
    print(extime/rep)
    print(sol2)
    results = np.append(results, extime/rep)
    results = np.append(results, [dk, db])
    results = np.append(results, np.array2string(sol2[:,0]))
    
    return results
    
```

```python
def incertidumbre1(om): # number of alternatives that would be explored with azzini   
    return np.sum(om.sum(axis=1) >= om.sum(axis=0)) # / om.shape[0]
def incertidumbre2(om): # sum in relation to condorcet   
    return ((om > 5).sum(axis=1)).sum() # / (((n)*(n+1))/2)
def incertidumbre3(om): # for each pair keep the max
    values = om[np.triu_indices(om.shape[0], k = 1)]
    votes = om[0,1]+om[1,0]
    #print(np.maximum(votes-values, values))
    return np.sum(np.maximum(votes-values, values))
def incertidumbre4(om): # ponderando por los puestos que acierta
    votes = om[0,1]+om[1,0]
    ranking = (om > 5).sum(axis=1)
    #print(votes)
    #print(ranking)
    count = 0
    for i in range(1,om.shape[0]+1):
        are = np.where(ranking == (i-1))[0]
        if are.size >= 1:
            #print("i = {}, adding {}".format(i-1, i /are.size))
            count += 1 / are.size
    return count
def incertidumbre5(om): # ponderando por los puestos que acierta
    votes = om[0,1]+om[1,0]
    ranking = (om > 5).sum(axis=1)
    #print(votes)
    #print(ranking)
    count = 0
    for i in range(1,om.shape[0]+1):
        are = np.where(ranking == (i-1))[0]
        if are.size >= 1:
            #print("i = {}, adding {}".format(i-1, i /are.size))
            count += i / are.size
    return count
```
